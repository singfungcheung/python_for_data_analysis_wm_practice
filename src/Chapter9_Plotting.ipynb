### Plots reside in the figure object.

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1) # 2x2 and 1st plot.
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

ax1.hist(np.random.standard_normal(100), bins=20, color="black", alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_normal(30))
ax3.plot(np.random.standard_normal(50).cumsum(), color="black", linestyle="dashed")
```

### Another convenient way of doing the subplots is to do plt.subplots.

```python
fig , axes = plt.subplots(2, 3)
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.standard_normal(500), bins=50, color="black", alpha=0.5)
fig.subplots_adjust(wspace=0, hspace=0)
```

### In the above graph, the graphs share the same x and y axis. The width and height space are 0s so they are squished together.
